```python
# Importando as bibliotecas necess√°rias para o projeto.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# ORGANIZANDO OS DADOS

### Criando o frame com os dados de Abril

```python
anp_abril = pd.read_csv('https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/shpc/dsan/2023/precos-gasolina-etanol-04.csv', sep = ';')
anp_abril.head(3)
```

### Adicionando a coluna de m√™s

```python
anp_abril['Mes'] = 'ABRIL'
anp_abril.head(3)
```

### Criando o frame com os dados de Mar√ßo

```python
anp_marco = pd.read_csv('https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/shpc/dsan/2023/precos-gasolina-etanol-03.csv', sep = ';')
anp_marco.head(3)
```

### Adicionando a coluna de mes

```python
anp_marco['Mes'] = 'MAR√áO'
anp_marco.head(3)
```

### Concatenando os dois frames

```python
relatorio = pd.concat([anp_abril,anp_marco], ignore_index= True)
relatorio.head(3)
```

### Substituindo a virgula por ponto e convertendo a coluna 'Valor de Venda' para float para ent√£o conseguir fazer as devidas an√°lises com os n√∫meros

```python
relatorio['Valor de Venda']= relatorio['Valor de Venda'].str.replace(',', '.')
relatorio['Valor de Venda'] = relatorio['Valor de Venda'].astype(float)
```

### Deletando a coluna 'Valor de Compra' pois a mesma se encontra totalmente vazia


```python
del relatorio['Valor de Compra']
```

### Separando os tipos de combust√≠veis para an√°lises futuras

```python
gasolina = relatorio.query('Produto == "GASOLINA"')
```

```python
etanol = relatorio.query('Produto == "ETANOL"')
```

```python
gasolina_aditivada = relatorio.query('Produto == "GASOLINA ADITIVADA"')
```

### Agrupando as datas por m√™s para fazer o gr√°fico de comportamento referente a cada m√™s e seu respectivo tipo de combust√≠vel

```python
etanol_abril = etanol.query('Mes == "ABRIL"')
```

```python
etanol_marco = etanol.query('Mes == "MAR√áO"')
```

```python
gasolina_abril = gasolina.query('Mes == "ABRIL"')
```

```python
gasolina_marco = gasolina.query('Mes == "MAR√áO"')
```

```python
aditivada_abril = gasolina_aditivada.query('Mes == "ABRIL"')
```

```python
aditivada_marco = gasolina_aditivada.query('Mes == "MAR√áO"')
```

```python
preco_etanol_abril = etanol_abril[['Data da Coleta', 'Valor de Venda']].groupby('Data da Coleta').mean().sort_values(by = 'Data da Coleta').round(2)
```

```python
preco_etanol_mar = etanol_marco[['Data da Coleta', 'Valor de Venda']].groupby('Data da Coleta').mean().sort_values(by = 'Data da Coleta').round(2)
```

```python
preco_gasolina_abril = gasolina_abril[['Data da Coleta', 'Valor de Venda']].groupby('Data da Coleta').mean().sort_values(by = 'Data da Coleta').round(2)
```

```python
preco_gasolina_marco = gasolina_marco[['Data da Coleta', 'Valor de Venda']].groupby('Data da Coleta').mean().sort_values(by = 'Data da Coleta').round(2)
```

```python
preco_aditivada_abril = aditivada_abril[['Data da Coleta', 'Valor de Venda']].groupby('Data da Coleta').mean().sort_values(by = 'Data da Coleta').round(2)
```

```python
preco_aditivada_marco = aditivada_marco[['Data da Coleta', 'Valor de Venda']].groupby('Data da Coleta').mean().sort_values(by = 'Data da Coleta').round(2)
```

### Etanol

```python
# criando vari√°veis que contenham uma lista com os valores de venda de suas respectivas regi√µes
N_Etanol = etanol.loc[etanol['Regiao - Sigla'] == 'N']['Valor de Venda'].tolist()
NE_Etanol = etanol.loc[etanol['Regiao - Sigla'] == 'NE']['Valor de Venda'].tolist()
SE_Etanol = etanol.loc[etanol['Regiao - Sigla'] == 'SE']['Valor de Venda'].tolist()
S_Etanol = etanol.loc[etanol['Regiao - Sigla'] == 'S']['Valor de Venda'].tolist()
CO_Etanol = etanol.loc[etanol['Regiao - Sigla'] == 'CO']['Valor de Venda'].tolist()

# Encontrar o valor m√°ximo de linhas entre as regi√µes
max_rows = max(len(N_Etanol), len(NE_Etanol), len(SE_Etanol), len(S_Etanol), len(CO_Etanol))

# Preencher os valores ausentes com NaNs para manter o alinhamento das colunas
N_Etanol += [float('nan')] * (max_rows - len(N_Etanol))
NE_Etanol += [float('nan')] * (max_rows - len(NE_Etanol))
SE_Etanol += [float('nan')] * (max_rows - len(SE_Etanol))
S_Etanol += [float('nan')] * (max_rows - len(S_Etanol))
CO_Etanol += [float('nan')] * (max_rows - len(CO_Etanol))

# Criar DataFrame com as listas preenchidas
reg_etanol = pd.DataFrame({'Norte': N_Etanol, 'Nordeste': NE_Etanol, 'Sudeste': SE_Etanol,
                           'Sul': S_Etanol, 'Centro Oeste': CO_Etanol})

reg_etanol.corr()

```

```python
# Configura√ß√£o do mapa de calor com escala de cores personalizada
plt.figure(figsize=(6, 5))
sns.heatmap(reg_etanol.corr(), annot=True, cmap='RdYlGn', vmin=-1, vmax=1)

# Configura√ß√µes adicionais
plt.title('Mapa de Calor da Correla√ß√£o  - Etanol')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()

# Exibi√ß√£o do mapa de calor
plt.show()
```
